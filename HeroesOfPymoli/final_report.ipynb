```python
# Dependencies and setup
import pandas as pd

# File path
file = "Resources/purchase_data.csv"

# Read file
df = pd.read_csv(file)

# Display
df.head()
```

```python
# Player count

# Calculation
player_count = len(df["SN"].unique())

# Dataframe to store calculation result
total_players_df = pd.DataFrame ({"Total Players": [player_count]})

# Display
total_players_df
```

```python
# Purchasing analysis (total)

# Calculation 
unique_items_count = len(df["Item Name"].unique())
average_price = df["Price"].mean()
count_purchase = df["Item Name"].count()
revenue = df["Price"].sum()

# Dataframe to store calculation results
purchase_analysis_df = pd.DataFrame({"Number of Unique Items":[unique_items_count],
                                    "Average Price":[average_price],
                                    "Number of Purchases":[count_purchase],
                                    "Total Revenue":[revenue]
                                    })

# Formatting
purchase_analysis_df ["Average Price"] = purchase_analysis_df ["Average Price"].map("${:,.2f}".format)
purchase_analysis_df ["Total Revenue"] = purchase_analysis_df ["Total Revenue"].map("${:,.2f}".format)

# Display
purchase_analysis_df
```

```python
# Gender demographics

# Dataframe remove SN duplicates and set gender as index
no_duplicate_df = df.drop_duplicates(subset=["SN"]).copy()
gender_index = no_duplicate_df.set_index ("Gender")

# Calculation
male_list = gender_index.loc["Male", "SN"]
male_count = male_list.count()
male_percentage = male_count / player_count
female_list = gender_index.loc["Female", "SN"]
female_count = female_list.count()
female_percentage = female_count / player_count
other_count = player_count - (male_count + female_count)
other_percentage = 1 - (male_percentage + female_percentage)

# Dataframe to store calculation results
gender_df = pd.DataFrame([{"Gender":"Male","Total Count": male_count, "Percentage of Players": male_percentage},
                        {"Gender":"Female","Total Count": female_count, "Percentage of Players": female_percentage},
                        {"Gender":"Other / Non-Disclosed", "Total Count": other_count, "Percentage of Players": other_percentage}
                        ])
gender_df = gender_df.set_index ("Gender")

# Formatting
gender_df ["Percentage of Players"] = gender_df ["Percentage of Players"].map("{:.2%}".format)
gender_df.index.name = None

# Display
gender_df
```

```python
# Purchasing analysis by gender

# Calculation
purchase_gender = df.groupby("Gender")
purchase_count = purchase_gender["Purchase ID"].count()
avg_purchase_price = purchase_gender["Price"].mean()
total_purchase_value = purchase_gender["Price"].sum()
avg_total_purchase_per_person = total_purchase_value / gender_df["Total Count"]

# Dataframe to store calculation results
purchase_bygender_df = pd.DataFrame({"Gender": ["Female","Male","Other / Non-Disclosed"],
                                    "Purchase Count": purchase_count,
                                    "Average Purchase Price": avg_purchase_price,
                                    "Total Purchase Value": total_purchase_value,
                                    "Avg Total Purchase per Person": avg_total_purchase_per_person
                                    })
purchase_bygender_df = purchase_bygender_df.set_index ("Gender")

# Formatting
purchase_bygender_df ["Average Purchase Price"] = purchase_bygender_df ["Average Purchase Price"].map("${:,.2f}".format)
purchase_bygender_df ["Total Purchase Value"] = purchase_bygender_df ["Total Purchase Value"].map("${:,.2f}".format)
purchase_bygender_df ["Avg Total Purchase per Person"] = purchase_bygender_df ["Avg Total Purchase per Person"].map("${:,.2f}".format)

# Display
purchase_bygender_df
```

```python
# Age Demographics

# Establishing bins
bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Categorizing data using the bins
no_duplicate_df ["Age Range"] = pd.cut(no_duplicate_df["Age"], bins, labels=group_names, include_lowest=True)

# Calculation
age_group = no_duplicate_df.groupby("Age Range")
total_count = age_group ["Age"].count()
percentage_of_players = age_group ["Age"].count() / player_count

# Dataframe to store calculation results
age_group_df = pd.DataFrame ({"Total Count": total_count,
                              "Percentage of Players": percentage_of_players})

# Formatting
age_group_df ["Percentage of Players"] = age_group_df ["Percentage of Players"].map("{:.2%}".format)
age_group_df.index.name = None

# Display
age_group_df
```

```python
# Purchasing Analysis by Age

# Categorizing data using the bins
df["Age Ranges"] = pd.cut(df["Age"], bins, labels=group_names, include_lowest=True)

# Calculation
age_group_purchase = df.groupby(["Age Ranges"])
purchase_count2 = age_group_purchase ["Age Ranges"].count()
avg_purchase_price2 = age_group_purchase ["Price"].mean()
total_purchase_value2 = age_group_purchase ["Price"].sum()
avg_total_purchase_per_person2 = age_group_purchase ["Price"].sum()/age_group_df["Total Count"]

# Dataframe to store calculation results
age_group_purchase_df = pd.DataFrame({"Purchase Count":purchase_count2,
                                      "Average Purchase Price":avg_purchase_price2, 
                                      "Total Purchase Value":total_purchase_value2,
                                      "Avg Total Purchase per Person":avg_total_purchase_per_person2
                                      })

# Formatting
age_group_purchase_df ["Average Purchase Price"] = age_group_purchase_df ["Average Purchase Price"].map("${:,.2f}".format)
age_group_purchase_df ["Total Purchase Value"] = age_group_purchase_df ["Total Purchase Value"].map("${:,.2f}".format)
age_group_purchase_df ["Avg Total Purchase per Person"] = age_group_purchase_df ["Avg Total Purchase per Person"].map("${:,.2f}".format)

# Display
age_group_purchase_df
```

```python
# Top Spenders

# Calculation
top_spender = df.groupby(["SN"])
purchase_count3 = top_spender ["Price"].count()
avg_purchase_price3 = top_spender["Price"].mean()
total_purchase_value3 = top_spender["Price"].sum()

# Dataframe to store the results
top_spender_df = pd.DataFrame({"Purchase Count": purchase_count3,
                               "Average Purchase Price": avg_purchase_price3,
                               "Total Purchase Value": total_purchase_value3                               
                               })

# Sorting total purchase value
sorted = top_spender_df.sort_values("Total Purchase Value",ascending=False)

# Formatting
sorted ["Average Purchase Price"] = sorted ["Average Purchase Price"].map("${:,.2f}".format)
sorted ["Total Purchase Value"] = sorted ["Total Purchase Value"].map("${:,.2f}".format)

# Display
sorted.head()
```

```python
# Most Popular Items

# Retrieve the Item ID, Item Name, and Item Price columns
item_data = df[["Item ID", "Item Name", "Price"]]

# Calculation
popular_item = item_data.groupby(["Item ID","Item Name"])
purchase_count4 = popular_item ["Price"].count()
item_price = popular_item["Price"].mean()
total_purchase_value4 = popular_item["Price"].sum()

# Dataframe to store the results
popular_item_df = pd.DataFrame({"Purchase Count": purchase_count4,
                               "Item Price": item_price,
                               "Total Purchase Value": total_purchase_value4                             
                               })

# Sorting by purchase count
sorted_popular_item = popular_item_df.sort_values("Purchase Count",ascending=False)

# Formatting
sorted_popular_item ["Item Price"] = sorted_popular_item ["Item Price"].map("${:,.2f}".format)
sorted_popular_item ["Total Purchase Value"] = sorted_popular_item ["Total Purchase Value"].map("${:,.2f}".format)

# Display
sorted_popular_item.head()
```

```python
# Most Profitable Items
 
# Sorting by total purchase value
sorted_profitable_item = popular_item_df.sort_values("Total Purchase Value",ascending=False)

# Formatting
sorted_profitable_item ["Item Price"] = sorted_profitable_item ["Item Price"].map("${:,.2f}".format)
sorted_profitable_item ["Total Purchase Value"] = sorted_profitable_item ["Total Purchase Value"].map("${:,.2f}".format)

# Display
sorted_profitable_item.head()
```

# Observable trends based on the data:

- Male players make up the majority of players, more than 80%, and bring more than 80% of the total revenue.
- On average male spend less per person and buy lower priced items compared to other gender categories.
- More than 60% of revenue is derived from players aged 15 – 24. This age range accounts for more than 60% of the players.
- Top 2 most popular and profitable items are Final Critic and Oathbreaker, Last Hope of the Breaking Storm.
